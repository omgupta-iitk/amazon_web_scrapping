```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def total_rating(topic_page,ind):
    total_ratings = []
    new1 =topic_page.find_all(class_ = 'a-size-small')
    temp = 0
    for tags in new1:
        if temp==50:
            break
        else:
            if len(ind)==0:
                if len(tags['class'])==1:
                    total_ratings.append(tags.text)
                    temp +=1
            else:
                if temp in ind:
                    total_ratings.append('NA')
                    temp +=1
                    continue
                if len(tags['class'])==1:
                    total_ratings.append(tags.text)
                    temp +=1
    return total_ratings
```

```python
def genre_scrap(topic_page):
    base_url = 'https://www.amazon.in'
    temp =0
    temp2 =topic_page.find_all('div',class_ ="p13n-sc-uncoverable-faceout")
    auth_name =[]
    auth_link_list = []
    rating =[]
    book_type =[]
    bk_name =[]
    total_ratings =[]
    price = []
    book_link = []
    count =0
    ind = []
    for tags in temp2:



        # Ratings
        try:
            t3 =tags.find('span', class_= 'a-icon-alt')
            rating.append(t3.text)
        except:
            rating.append('NA')
            ind.append(count)
        count+=1



        # Author name and Author Link
        try:
            t1 =tags.find('a', class_="a-size-small a-link-child").find('div')
            t2 = tags.find('a',class_ ='a-size-small a-link-child')
            auth_link_list.append(base_url + t2['href'])
            auth_name.append(t1.text)
        except:
            auth_link_list.append('NA')
            try:
                t =tags.find('span',class_ ="a-size-small a-color-base")
                auth_name.append(t.text)
            except:
                auth_name.append('NA')
        


        # BOOK Type
        t5 = tags.find(class_ = 'a-size-small a-color-secondary a-text-normal')
        book_type.append(t5.text)
        # Book Name
        t6 = tags.find('img', {'class': "a-dynamic-image p13n-sc-dynamic-image p13n-product-image"})
        bk_name.append(t6['alt'])



        # price
        try:
            t7 = tags.text.index('â‚¹')
            price.append(tags.text[t7+1:-1])
        except:
            price.append('NA')


    
        # BOOk Link
        t8 = tags.find('a',class_ ='a-link-normal')["href"]
        book_link.append(base_url + t8)
    


    
    # Total Ratings
    total_ratings = total_rating(topic_page,ind)


    dict = {'Book Name': bk_name,
            'Author Name':auth_name,
            'Rating':rating,
            'Total Ratings': total_ratings,
            'Book Type': book_type,
            'Price': price,
            'Book Link': book_link,
            'Author Link': auth_link_list       
            }

    return dict,count

```

```python
page = requests.get('https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_unv_books_1_1318158031_2')
page = BeautifulSoup(page.text, 'html.parser')
genre_tags = page.find_all('div', class_ ="_p13n-zg-nav-tree-all_style_zg-browse-item__1rdKf _p13n-zg-nav-tree-all_style_zg-browse-height-large__1z5B8")
base_url = 'https://www.amazon.in'


for tags in genre_tags:
        if tags.text == 'Books':
                continue
        count =0
        while count==0:
            str_tags = tags.find('a')['href']
            genre = tags.text.strip()
            genre_url = (base_url + str_tags)
            topic_page = requests.get(genre_url)
            topic_page = BeautifulSoup(topic_page.text, 'html.parser')
            dict,count = genre_scrap(topic_page)
            data = pd.DataFrame(dict)   
        data.to_csv('{}.csv'.format(genre),index=False)
        print(genre+' csv is created successfully')
```
